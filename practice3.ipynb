```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the staircase function below.
def staircase(n):
    a=''
    for _ in range(n):
        a+='#'
        print(a.rjust(n,' '))

if __name__ == '__main__':
    n = int(input())

    staircase(n)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys
import itertools

# Complete the miniMaxSum function below.
def miniMaxSum(arr):
    arr_sum = []
    arr1 = itertools.combinations(arr,4)
    for i in list(arr1):
        arr_sum.append(sum(i))
    arr_sum.sort()
    print(arr_sum[0],arr_sum[-1])

if __name__ == '__main__':
    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```

```python
def birthdayCakeCandles(ar):
    ar.sort()
    tall = ar.pop()
    if 1<=tall<=10**7:
        cnt = 1
        while(ar):
            if 1<= ar[-1] <=10**7: 
                if ar.pop() == tall:
                    cnt+=1
                else:
                    break
    return(cnt)

if __name__ == '__main__':
    a = int(input())
    ar = map(int,input().split())
    print(birthdayCakeCandles(list(ar)))
```
